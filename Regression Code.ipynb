```python
DataFrame = pd.read_csv('updated_12_07.csv')

# Organizing the AQI 
AQI_Sort = DataFrame.sort_values(['90th Percentile AQI']) # Sorts the AQI according towards Median AQI 
AQI_Sort = AQI_Sort.reset_index(drop=True)

# Separating the 90th Percentile AQI into Good, Moderate,Lightly Polluted, Moderately Pollued, and Hazardous Air Quality According to EPA Standards
DataFrame.loc[DataFrame['90th Percentile AQI'] <=50, 'Quality'] = 'Good'
DataFrame.loc[(DataFrame['90th Percentile AQI'] >= 51) & (DataFrame['90th Percentile AQI'] <= 100),'Quality'] = 'Moderate'
DataFrame.loc[(DataFrame['90th Percentile AQI'] >= 101) & (DataFrame['90th Percentile AQI'] <= 150),'Quality'] = 'Lightly Polluted'
DataFrame.loc[(DataFrame['90th Percentile AQI'] >= 151) & (DataFrame['90th Percentile AQI'] <= 200),'Quality'] = 'Moderately Polluted'
DataFrame.loc[DataFrame['90th Percentile AQI'] > 201, 'Quality'] = 'Hazardous'
data = DataFrame['Quality'].value_counts().to_frame('counts')
DataFrame.to_csv('update.csv') # double check the .CSV

Good = DataFrame[DataFrame['Quality'].str.contains('Good')]
Moderate = DataFrame[DataFrame['Quality'].str.contains('Moderate')]
Lightly_Polluted = DataFrame[DataFrame['Quality'].str.contains('Lightly Polluted')]
Moderately_Polluted = DataFrame[DataFrame['Quality'].str.contains('Moderately Polluted')]
Hazardous = DataFrame[DataFrame['Quality'].str.contains('Hazardous')]


# Visualizing the AQI Data 
# plotting to log scale for population 
pop_log_good = np.log10(Good['Population (2014-2020)'])
pop_log_mod = np.log10(Moderate['Population (2014-2020)'])
pop_log_lightpol = np.log10(Lightly_Polluted['Population (2014-2020)'])
pop_log_modpol = np.log10(Moderately_Polluted['Population (2014-2020)'])
pop_log_haz = np.log10(Hazardous['Population (2014-2020)'])


fig=go.Figure()

fig.add_trace(go.Scatter(y = Good['Quality'],
                                x = pop_log_good,
                                mode='markers',
                                name= 'Good AQI'))
                                

fig.add_trace(go.Scatter(y = Moderate['Quality'],
                                x = pop_log_mod,
                                mode='markers',
                                name= 'Moderate AQI'))
                                
              
fig.add_trace(go.Scatter(y = Lightly_Polluted['Quality'],
                                x = pop_log_lightpol,
                                mode='markers',
                                name= 'Lightly Polluted'))

fig.add_trace(go.Scatter(y = Moderately_Polluted['Quality'],
                                x = pop_log_modpol,
                                mode='markers',
                                name= 'Moderately Polluted'))

fig.add_trace(go.Scatter(y = Hazardous['Quality'],
                                x = pop_log_haz,
                                mode='markers',
                                name= 'Hazardous'))
                                
fig.update_layout(
    title="AQI Vs. Population",
    xaxis_title="Population",
    yaxis_title="AQI Quality"
    )
array = ['Good','Moderate','Lightly Polluted','Moderately Polluted','Hazardous']
ig.update_yaxes(categoryarray=<VALUE>)
```
